```python
import sqlite3
import pandas as pd

# Load CSV into DataFrame
df = pd.read_csv('/content/Superstore.csv', encoding='latin-1')
```

```python
# Connect to SQLite (in-memory or file-based)
conn = sqlite3.connect('superstore.db')
cursor = conn.cursor()

# Store DataFrame into SQLite table
df.to_sql('superstore', conn, index=False, if_exists='replace')
```

```python
# List all tables in the SQLite database
tables_query = "SELECT name FROM sqlite_master WHERE type='table';"
pd.read_sql_query(tables_query, conn)
```

```python
# Use PRAGMA to view table structure for 'superstore'
pd.read_sql_query("PRAGMA table_info(superstore);", conn)
```

```python
# Task 2 - Simulate SELECT queries

# 1. SELECT * FROM customers --> For Superstore, this will be full table view
select_all = pd.read_sql_query("SELECT * FROM superstore LIMIT 5;", conn)

# 2. SELECT first_name, last_name FROM employees --> Simulated with Customer Name
select_columns = pd.read_sql_query("SELECT [Customer Name], [Segment] FROM superstore LIMIT 5;", conn)

select_all, select_columns
```

```python
# Task 3 - Filtering with WHERE conditions

# 1. Sales greater than 100
pd.read_sql_query("SELECT * FROM superstore WHERE Sales > 100 LIMIT 5;", conn)
```

```python
# 2. Discount between 0.1 and 0.3
pd.read_sql_query("SELECT * FROM superstore WHERE Discount BETWEEN 0.1 AND 0.3 LIMIT 5;", conn)
```

```python
# 3. State NOT equal to 'California'
pd.read_sql_query("SELECT * FROM superstore WHERE State != 'California' LIMIT 5;", conn)
```

```python
# 4. Product Name contains 'Chair'
pd.read_sql_query("SELECT * FROM superstore WHERE [Product Name] LIKE '%Chair%' LIMIT 5;", conn)
```

```python
# 5. Region IN a list
pd.read_sql_query("SELECT * FROM superstore WHERE Region IN ('West', 'South') LIMIT 5;", conn)
```

```python
#Task 4: Sorting and Limiting
df_asc = pd.read_sql_query("SELECT * FROM superstore ORDER BY Sales ASC;", conn)
df_asc.head()
```

```python
df_desc = pd.read_sql_query("SELECT * FROM superstore ORDER BY Sales DESC;", conn)
df_desc.head()
```

```python
df_low5 = pd.read_sql_query("SELECT * FROM superstore ORDER BY Sales ASC LIMIT 5;", conn)
df_low5
```

```python
# Task 5: Basic Aggregations using SQL on Superstore

# 1. Basic aggregation: COUNT, AVG, SUM
pd.read_sql_query("""
SELECT
    COUNT(*) AS total_orders,
    AVG(Sales) AS avg_sales,
    SUM(Sales) AS total_sales
FROM superstore;
""", conn)

```

```python
# 2. Grouped aggregation: total sales by Region
pd.read_sql_query("""
SELECT
    Region,
    COUNT(*) AS num_orders,
    AVG(Sales) AS avg_sales,
    SUM(Sales) AS total_sales
FROM superstore
GROUP BY Region;
""", conn)
```
