```python
# Group by type and get average release year
df.groupby('type')['release_year'].mean().reset_index()
```

```python
# Group by country: count of titles
df.groupby('country')['title'].agg(['count']).reset_index().head()
```

```python
# Group by country: min, max, mean release years
df.groupby('country')['release_year'].agg(['min', 'max', 'mean']).reset_index().head()

```

```python
# Sample 1
shows_info = df[['show_id', 'title', 'type', 'release_year']]
```

```python
# Sample 2:
origin_info = pd.DataFrame({
    'show_id': ['s1', 's2', 's5', 's10'],  # example IDs
    'origin_platform': ['Netflix', 'Netflix', 'Hulu', 'Amazon Prime']
})
```

```python
# INNER JOIN: Only shows present in both
pd.merge(shows_info, origin_info, on='show_id', how='inner')
```

```python
# LEFT JOIN: All Netflix shows, match platform if exists
pd.merge(shows_info, origin_info, on='show_id', how='left')
```

```python
# OUTER JOIN: All show IDs from both DataFrames
pd.merge(shows_info, origin_info, on='show_id', how='outer')
```
